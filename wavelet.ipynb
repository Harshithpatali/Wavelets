```python
# =====================================================
# 📈 Sensex Forecasting using LSTM & Haar Wavelet Model
# Author: Harshith | Date: October 2025
# =====================================================

# ===============================
# 📦 Import All Dependencies
# ===============================

# Data manipulation
import numpy as np
import pandas as pd
import yfinance as yf
import datetime as dt

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-v0_8-darkgrid')

# Machine learning
from sklearn.model_selection import train_test_split, TimeSeriesSplit
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.ensemble import RandomForestRegressor

# Deep learning (LSTM)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# Prophet model
from prophet import Prophet

# Wavelet transforms
import pywt

# Utility
import warnings
warnings.filterwarnings('ignore')

print("✅ All dependencies imported successfully (without pmdarima)!")

# ============================================
# 📊 STEP 2: Download and Clean Sensex Data
# ============================================

ticker = "^BSESN"   # Yahoo Finance ticker for SENSEX
start_date = "2000-01-01"
end_date = dt.datetime.today().strftime('%Y-%m-%d')

# Download data
sensex = yf.download(ticker, start=start_date, end=end_date, progress=False)

# Flatten MultiIndex if needed
if isinstance(sensex.columns, pd.MultiIndex):
    sensex.columns = [col[0] for col in sensex.columns]

print(f"✅ Downloaded {len(sensex)} records of Sensex data from {start_date} to {end_date}")

# Reset index
sensex.reset_index(inplace=True)

# Rename columns
sensex.rename(columns={
    "Date": "date",
    "Open": "open",
    "High": "high",
    "Low": "low",
    "Close": "close",
    "Adj Close": "adj_close",
    "Volume": "volume"
}, inplace=True)

# Drop duplicates and handle missing values
sensex.drop_duplicates(subset=["date"], inplace=True)
sensex.dropna(inplace=True)

# Add derived features
sensex["returns"] = sensex["close"].pct_change() * 100
sensex["log_returns"] = np.log1p(sensex["returns"])
sensex["year"] = sensex["date"].dt.year
sensex["month"] = sensex["date"].dt.month
sensex["day"] = sensex["date"].dt.day
sensex.dropna(inplace=True)

print("\n📋 Cleaned Data Preview:")
display(sensex.head())

# ============================================
# 📉 STEP 3: Data Preparation for Modeling
# ============================================

data = sensex[['date', 'close']].copy()
data.set_index('date', inplace=True)

# Scale the close price between 0 and 1
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# Function to create sequences for LSTM
def create_sequences(dataset, time_step=60):
    X, y = [], []
    for i in range(len(dataset) - time_step - 1):
        X.append(dataset[i:(i + time_step), 0])
        y.append(dataset[i + time_step, 0])
    return np.array(X), np.array(y)

time_step = 60
X, y = create_sequences(scaled_data, time_step)

# Reshape input to [samples, time_steps, features]
X = X.reshape(X.shape[0], X.shape[1], 1)

# Split into train/test sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

print(f"✅ Data prepared for LSTM — Train size: {len(X_train)}, Test size: {len(X_test)}")

# ============================================
# 🤖 STEP 4: Baseline LSTM Model
# ============================================

lstm_model = Sequential([
    LSTM(100, return_sequences=True, input_shape=(X_train.shape[1], 1)),
    Dropout(0.2),
    LSTM(100, return_sequences=False),
    Dropout(0.2),
    Dense(50, activation='relu'),
    Dense(1)
])

lstm_model.compile(optimizer='adam', loss='mean_squared_error')

early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history = lstm_model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=50,
    batch_size=32,
    callbacks=[early_stop],
    verbose=1
)

# Plot training performance
plt.figure(figsize=(10,5))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("LSTM Training & Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()

print("✅ Baseline LSTM model trained successfully!")

# ============================================
# 📊 STEP 5: Evaluate Baseline LSTM
# ============================================

predicted = lstm_model.predict(X_test)
predicted_prices = scaler.inverse_transform(predicted)
actual_prices = scaler.inverse_transform(y_test.reshape(-1, 1))

baseline_rmse = np.sqrt(mean_squared_error(actual_prices, predicted_prices))
baseline_mae = mean_absolute_error(actual_prices, predicted_prices)
baseline_r2 = r2_score(actual_prices, predicted_prices)

print(f"✅ Baseline LSTM RMSE: {baseline_rmse:.2f}, MAE: {baseline_mae:.2f}, R²: {baseline_r2:.6f}")

plt.figure(figsize=(10,6))
plt.plot(actual_prices, label='Actual')
plt.plot(predicted_prices, label='Predicted')
plt.title("Baseline LSTM: Actual vs Predicted Sensex")
plt.legend()
plt.show()

# ============================================
# 🌊 STEP 6: Wavelet Denoising + Wavelet-LSTM
# ============================================

def wavelet_denoise(data, wavelet="haar", level=1):
    coeffs = pywt.wavedec(data.squeeze(), wavelet, level=level)
    coeffs[1:] = [np.zeros_like(detail) for detail in coeffs[1:]]
    return pywt.waverec(coeffs, wavelet).reshape(-1, 1)

# Apply Haar wavelet denoising
denoised_data = wavelet_denoise(scaled_data, wavelet="haar", level=1)

# Recreate sequences
Xw, yw = create_sequences(denoised_data, time_step)
Xw = Xw.reshape(Xw.shape[0], Xw.shape[1], 1)

train_size_w = int(len(Xw) * 0.8)
Xw_train, Xw_test = Xw[:train_size_w], Xw[train_size_w:]
yw_train, yw_test = yw[:train_size_w], yw[train_size_w:]

# Build and train Wavelet-LSTM
wavelet_lstm_model = Sequential([
    LSTM(100, return_sequences=True, input_shape=(Xw_train.shape[1], 1)),
    Dropout(0.2),
    LSTM(100, return_sequences=False),
    Dropout(0.2),
    Dense(50, activation='relu'),
    Dense(1)
])

wavelet_lstm_model.compile(optimizer='adam', loss='mean_squared_error')

history_w = wavelet_lstm_model.fit(
    Xw_train, yw_train,
    validation_data=(Xw_test, yw_test),
    epochs=50,
    batch_size=32,
    callbacks=[early_stop],
    verbose=1
)

print("✅ Wavelet-LSTM model trained successfully!")

# ============================================
# 📈 STEP 7: Evaluate & Compare Both Models
# ============================================

predicted_w = wavelet_lstm_model.predict(Xw_test)
predicted_w_prices = scaler.inverse_transform(predicted_w)
actual_w_prices = scaler.inverse_transform(yw_test.reshape(-1, 1))

wavelet_rmse = np.sqrt(mean_squared_error(actual_w_prices, predicted_w_prices))
wavelet_mae = mean_absolute_error(actual_w_prices, predicted_w_prices)
wavelet_r2 = r2_score(actual_w_prices, predicted_w_prices)

comparison = pd.DataFrame({
    'Model': ['Baseline LSTM', 'Wavelet-LSTM (Haar)'],
    'RMSE': [baseline_rmse, wavelet_rmse],
    'MAE': [baseline_mae, wavelet_mae],
    'R2': [baseline_r2, wavelet_r2]
})

print("\n📊 Final Model Comparison:")
display(comparison)

plt.figure(figsize=(10,6))
plt.plot(actual_prices[-500:], label='Actual')
plt.plot(predicted_prices[-500:], label='Baseline LSTM')
plt.plot(predicted_w_prices[-500:], label='Wavelet-LSTM (Haar)')
plt.title("Sensex Forecast Comparison (Last 500 Days)")
plt.legend()
plt.show()

# ============================================
# 🧾 STEP 8: Save Models & Report
# ============================================

import json

lstm_model.save("baseline_lstm_model.h5")
wavelet_lstm_model.save("wavelet_lstm_model.h5")

metrics = {
    "Baseline LSTM": {
        "RMSE": float(baseline_rmse),
        "MAE": float(baseline_mae),
        "R2": float(baseline_r2)
    },
    "Wavelet-LSTM (Haar)": {
        "RMSE": float(wavelet_rmse),
        "MAE": float(wavelet_mae),
        "R2": float(wavelet_r2)
    }
}

with open("model_comparison_metrics.json", "w") as f:
    json.dump(metrics, f, indent=4)

improvement = (1 - wavelet_rmse / baseline_rmse) * 100
print(f"\n🚀 Wavelet-LSTM improved RMSE by {improvement:.2f}% over baseline!")

print("✅ All models and metrics saved successfully!")
```

```python
from IPython.display import Markdown, display

display(Markdown(r"""
# 🧮 Mathematical Foundation: Haar Wavelet Transform

## 1️⃣ Introduction

The **Haar Wavelet Transform** is the simplest and oldest type of wavelet transform, introduced by **Alfréd Haar (1909)**.  
It converts a time-domain signal (like stock prices) into a representation that shows **both time and frequency information**, enabling **noise reduction and trend detection**.

While the **Fourier Transform** gives frequency information but loses time localization,  
**Wavelet Transform** preserves both — making it perfect for *non-stationary financial data* such as Sensex.

---

## 2️⃣ Haar Wavelet Basics

### a) Haar Scaling and Wavelet Functions

The Haar wavelet uses two simple basis functions:

$$
\phi(t) =
\begin{cases}
1, & 0 \le t < 1 \\
0, & \text{otherwise}
\end{cases}
$$

$$
\psi(t) =
\begin{cases}
1, & 0 \le t < \frac{1}{2} \\
-1, & \frac{1}{2} \le t < 1 \\
0, & \text{otherwise}
\end{cases}
$$

Here:
- $\phi(t)$ is the **scaling function** (represents average or trend)  
- $\psi(t)$ is the **wavelet function** (represents detail or noise)

---

## 3️⃣ Discrete Haar Wavelet Transform (DWT)

For a signal $S = [s_1, s_2, s_3, s_4, \ldots]$,  
the DWT computes **average (approximation)** and **difference (detail)** coefficients as:

$$
A_i = \frac{s_{2i-1} + s_{2i}}{\sqrt{2}}
$$

$$
D_i = \frac{s_{2i-1} - s_{2i}}{\sqrt{2}}
$$

- $A_i$ → Approximation coefficients (low frequency → trend)  
- $D_i$ → Detail coefficients (high frequency → noise)

The inverse transform reconstructs the signal:
$$
s_{2i-1} = \frac{A_i + D_i}{\sqrt{2}}, \quad s_{2i} = \frac{A_i - D_i}{\sqrt{2}}
$$

---

## 4️⃣ Application in Financial Time Series

In stock data (like Sensex), high-frequency movements often represent **noise or volatility spikes**,  
while low-frequency components carry **meaningful long-term trends**.

By applying **Haar DWT**, we decompose the closing prices as:

$$
\text{Close}(t) = A_1(t) + D_1(t) + D_2(t) + \ldots + D_n(t)
$$

We can then **filter out** high-frequency components $D_k$ to denoise the data:
$$
\text{Cleaned Signal} = A_1(t)
$$

---

## 5️⃣ Why Haar for LSTM?

✅ **Advantages:**
- Computationally simple (integer operations)  
- Perfect for real-time filtering  
- Reduces overfitting by removing high-frequency noise  
- Keeps key price patterns for trend learning  

🧠 The LSTM then learns from a **smoothed, noise-reduced signal**, improving prediction accuracy and stability.

---

## 6️⃣ Visualization Example

A Haar decomposition separates the Sensex close prices into:

- **Approximation ($A_1$):** Smooth long-term price movement  
- **Detail ($D_1$):** Short-term fluctuations and market noise  

This denoised signal is fed into the Wavelet-LSTM model for superior forecasting results.

---

📘 **Summary:**  
The Haar Wavelet Transform performs multi-resolution analysis, allowing LSTM models to focus on informative long-term structures rather than noisy short-term variations — leading to improved generalization and predictive performance in financial forecasting.
"""))

```

```python
# ============================================
# 🧠 Visualization: LSTM Cell Architecture
# ============================================

import matplotlib.pyplot as plt
import matplotlib.patches as patches

plt.figure(figsize=(12, 6))
ax = plt.gca()

# Draw rectangles for gates
def add_gate(ax, xy, text, color):
    rect = patches.Rectangle(xy, 1.5, 1, edgecolor='black', facecolor=color, lw=1.5)
    ax.add_patch(rect)
    ax.text(xy[0] + 0.75, xy[1] + 0.5, text, ha='center', va='center', fontsize=11, color='black')

# Add LSTM gates
add_gate(ax, (0, 2), "Forget Gate\n$f_t$", "#F9D5A7")
add_gate(ax, (2, 2), "Input Gate\n$i_t$", "#A1DFFB")
add_gate(ax, (4, 2), "Candidate\n$\\tilde{C}_t$", "#C4E5AF")
add_gate(ax, (6, 2), "Output Gate\n$o_t$", "#FBC4C4")

# Add arrows for flow
ax.arrow(-0.5, 2.5, 0.5, 0, head_width=0.15, head_length=0.15, fc='k', ec='k')
ax.arrow(1.5, 2.5, 0.5, 0, head_width=0.15, head_length=0.15, fc='k', ec='k')
ax.arrow(3.5, 2.5, 0.5, 0, head_width=0.15, head_length=0.15, fc='k', ec='k')
ax.arrow(5.5, 2.5, 0.5, 0, head_width=0.15, head_length=0.15, fc='k', ec='k')

# Add input and output labels
ax.text(-0.8, 2.5, "$h_{t-1}, x_t$", va='center', fontsize=12)
ax.text(7.8, 2.5, "$h_t$", va='center', fontsize=12)

# Add memory cell
ax.add_patch(patches.Rectangle((3, 0.5), 1.5, 1, edgecolor='black', facecolor='#F0E68C', lw=1.5))
ax.text(3.75, 1.0, "Cell State\n$C_t$", ha='center', va='center', fontsize=11)

# Add connections between gates and cell state
ax.arrow(0.75, 2, 0, -0.4, head_width=0.1, head_length=0.1, fc='k', ec='k', linestyle='--', alpha=0.7)
ax.arrow(2.75, 2, 0, -0.4, head_width=0.1, head_length=0.1, fc='k', ec='k', linestyle='--', alpha=0.7)
ax.arrow(4.75, 2, 0, -0.4, head_width=0.1, head_length=0.1, fc='k', ec='k', linestyle='--', alpha=0.7)
ax.arrow(6.75, 2, 0, -0.4, head_width=0.1, head_length=0.1, fc='k', ec='k', linestyle='--', alpha=0.7)

# Add mathematical equations
equation_text = r"$f_t = \sigma(W_f \cdot [h_{t-1}, x_t] + b_f)$" + "\n" + \
                r"$i_t = \sigma(W_i \cdot [h_{t-1}, x_t] + b_i)$" + "\n" + \
                r"$\tilde{C}_t = \tanh(W_C \cdot [h_{t-1}, x_t] + b_C)$" + "\n" + \
                r"$o_t = \sigma(W_o \cdot [h_{t-1}, x_t] + b_o)$" + "\n" + \
                r"$C_t = f_t \cdot C_{t-1} + i_t \cdot \tilde{C}_t$" + "\n" + \
                r"$h_t = o_t \cdot \tanh(C_t)$"

ax.text(0.5, 3.8, "LSTM Gate Equations:", fontsize=10, weight='bold')
ax.text(0.5, 3.3, equation_text, fontsize=8, va='top')

# Title and cleanup
ax.set_xlim(-1.5, 9)
ax.set_ylim(0, 4)
ax.axis('off')
plt.title("🧠 LSTM Cell Architecture - Gate Flow Overview", fontsize=14, pad=10)
plt.tight_layout()
plt.show()

print("✅ LSTM architecture visualization generated successfully!")
```
